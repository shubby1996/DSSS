```python
from skimage import io
import copy

from numpy import fliplr, flipud
from skimage.draw import set_color
```

```python


class ImageDataManipulator():
    def __init__(self, url):
        self.image = io.imread(url)
        self.man_image = copy.deepcopy(self.image)
        
    def flip(self, axis):
        if(axis==0):
            print("Flip Horizontally")
            self.man_image = flipud(self.man_image)
        else:
            print("Flip Vertically")
            self.man_image = fliplr(self.man_image)
    
    def crop(self, x, y, height, width):
        max_height = self.man_image.shape[0]
        max_width = self.man_image.shape[1]
        if(height>max_height):
            height=max_height
        if(width>max_width):
            width=max_width
        if(x<= max_height and y<=max_width):
            self.man_image = self.man_image[x:x+height,y:y+width]
        else:
            self.man_image = self.man_image
        
    def occlude(self, x, y, height, width, fill):
        if y > self.man_image.shape[0] or x > self.man_image.shape[1]:
            print("Error: x or y bigger than the image dimensions")
            return
        
        if height > self.man_image.shape[0] - y:
            height = self.man_image.shape[0] - y
        if width > self.man_image.shape[1] - x:
            width = self.man_image.shape[1] - x
            
        if fill == 'white':
            self.man_image[y:y+height, x:x+width, ...] = 255 * np.ones(shape=(height, width, 3))
        elif fill == 'black':
            self.man_image[y:y+height, x:x+width, ...] = 255 * np.zeros(shape=(height, width, 3))
        elif fill == 'uniform':
            self.man_image[y:y+height, x:x+width, ...] = 255 * np.random.uniform(size=(height, width, 3))
        else:
            print("Parameter 'fill' not valid.")
            return
    
    def reset_manipulations(self):
        self.man_image = copy.deepcopy(self.image)
    
    def show_manipulated_image(self):
        plt.imshow(self.man_image)
        plt.show()
    
    def show_original_image(self):
        plt.imshow(self.image)
        plt.show()
```

```python
import matplotlib.pyplot as plt
url = 'https://www.fau.de/files/2019/07/Kollegienhaus_Malter_3-480x284.jpg'

my_manipulator = ImageDataManipulator(url)
# print(my_manipulator.image.shape)
my_manipulator.show_original_image()
```
